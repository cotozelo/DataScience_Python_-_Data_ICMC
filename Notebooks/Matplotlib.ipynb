##Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Exemplos .plot

### Dados

```python
empresa_a = [1000, 1500, 1200, 3000, 2100, 2000]
empresa_b = [750, 2300, 1000, 500, 1200, 1500]
```

### Gr√°fico simples

```python
#Grafico simples
plt.plot(empresa_a)
#ref https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot
```

### Figurando o tampo do gr√°fico

```python
#figsize=(largura, altura)
plt.figure(figsize=(10, 6))
plt.plot(empresa_a); #; retira a linha de informa√ß√£o [<matplotlib.lines.Line2D at 0x7f84652f4dd0>]
```

### Aplicando marca√ß√µes nos pontos

```python
plt.figure(figsize=(10, 6))
plt.plot(empresa_a, marker="o"); 
# podemos usar varios tipos de marker, '-', '+', 'x', ...
# ref https://matplotlib.org/stable/api/markers_api.html?highlight=marker
```

### Mudando o tipo de linha

```python
plt.figure(figsize=(10, 6))
plt.plot(empresa_a, marker="o", linestyle='--'); 
#linestyle muda o tempo de linha, pode usar varios tipo, '--', '-'
# ref https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html?highlight=linestyle
```

### Cor da linha

```python
plt.figure(figsize=(10, 6))
plt.plot(empresa_a, marker="o", linestyle='--', color='green'); 
# color define direntes tipos de cores
```

### T√≠tulos

```python
plt.figure(figsize=(10, 6))
plt.plot(empresa_a, '--o', color='green') # marker="o", linestyle='--' pode ser torcado por '--o'
plt.title("Lucro da empresa A")
plt.xlabel("M√™s")
plt.ylabel("Valores em R$")
plt.show()
```

### Duas linhas no mesmo gr√°fico

```python
plt.figure(figsize=(10, 6))
plt.plot(empresa_a, '--o', color='green')
plt.plot(empresa_b, '-x', color='blue')
plt.title("Lucro das empresas A e B")
plt.xlabel("M√™s")
plt.ylabel("Valores em R$")
plt.show()
```

### Legenda

ref https://matplotlib.org/stable/api/legend_api.html?highlight=legend#module-matplotlib.legend

```python
plt.figure(figsize=(10, 6))
plt.plot(empresa_a, '--o', color='green', label='Empresa A')
plt.plot(empresa_b, '-x', color='blue', label='Empresa B')
plt.title("Lucro das empresas A e B")
plt.xlabel("M√™s")
plt.ylabel("Valores em R$")
plt.legend(loc='best')
plt.show()
```
